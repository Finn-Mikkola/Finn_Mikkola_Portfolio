```python
import nltk
import random
import numpy as np
```

```python
class FalloutProgramming:
    def __init__(self):
        pass
    
    #Generate the words and characters
    def generate_lines(self, length=20, word_length=8):
        
        #Setting up the random words as a global variable
        global random_words
        
        #Downloading random words of a specific length so the game is different everytime
        nltk.download('words')
        word_list = nltk.corpus.words.words()
        w = [word.upper() for word in word_list if len(word) == word_length]
        random_words = random.sample(w, 20)
        
        #Characters to fill in the gaps
        chars = ['@', '{', '}', '[', ']', '$', '&', '^', '.', '*', '(', ')']
        
        #Print loading message
        print('\n\n Loading in user(f)\n\n')
        
        
        #Careful indexing for having the right number characters per line.
        count = 0
        for i in range(20):
            output = ''
            current_char_printed = 0
            diff = length - word_length
            rand_length = random.randint(0, diff)
            current_char_printed += rand_length
            
            char_string = np.random.choice(chars, size=rand_length, replace=True)
            output = ''.join(char_string)
        
            output += ''.join(random_words[count])
        
            current_char_printed += word_length
            remainder = length - current_char_printed
            char_string = np.random.choice(chars, size=remainder, replace=True)
            output += ''.join(char_string)
            count += 1
            print(output)
            
    def calculate_similarity(self, word1, word2):
        #Check if the length is right first
        if len(word1) != len(word2):
            print("Error watch your spelling")
            raise ValueError("Words must be of equal length")
        
        #Fast way to calculate similarity between two words
        similarity = sum(1 for char1, char2 in zip(word1, word2) if char1 == char2)
        print(f"Similarity for '{word1}': {similarity}/{len(word1)}")
        
    #Run the game, user gets 4 tries before they are locked out like in the game
    def play_game(self):
        fails = 0
        word = random_words[random.randint(0, 9)]
        # uncomment below to see the word printed with the possible words
        # print(word)
        count = 1
        
        #While loop to end once fails is greater than 3
        #Logic to determine if the user wins or loses
        while fails < 4:
            guess = str(input('\nPlease Enter Password:').upper())
            if guess in random_words:
                print(f'Attempt {count}: {guess}')
                self.calculate_similarity(guess, word)
                if guess == word:
                    print('\nSuccess! Access Granted')
                    return
                    break
            else:
                print('Error')
                print(f'Attempt {count}: {guess}')
            fails += 1
            count += 1
            
        #print lockout message
        print('-' * 22)
        print('\n----- LOCKED OUT -----\n')
        print('-' * 22)

# Example usage:
game = FalloutProgramming()
game.generate_lines()
game.play_game()
```
